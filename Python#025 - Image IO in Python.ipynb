### Python Image Library (PIL)

```python
from PIL import Image
```

```python
im = Image.open('Food.jpg')
```

```python
im.show()
```

```python
im.save('PIL_imagesave.jpg')
```

### Scikit-image

```python
from skimage.io import imread, imshow, imsave
```

```python
imSk = imread('Food.jpg')
```

```python
imshow(imSk)
```

```python
imsave('scikit_image.jpg', imSk)
```

### OpenCV

```python
import cv2
```

```python
imCv = cv2.imread('Food.jpg')
```

```python
plt.imshow(cv2.cvtColor(imCv, cv2.COLOR_BGR2RGB))
```

```python
cv2.imshow('Food Image', imCv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
