**LIBRARIES**

```python
import numpy as np                           # get the maximum value
from skimage.io import imread                # open the input image
from skimage.color import rgb2lab, lab2rgb   # convert the input image into Lab color space
import matplotlib.pyplot as plt              # plot the result 
```

**Input Image**

```python
im = imread('Lenna.png')
```

```python
plt.title('Original Image')
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.show()
```

```python
im.shape
```

(height, widht, colorChannels)
* `im[:,:,0]`: Read
* `im[:,:,1]`: Green
* `im[:,:,2]`: Blue

**Convert the Image Color Space: `skimage.color.rgb2lab`**

Conversion from RGB color space into CIELAB color space.

```python
# RGB --> Lab --> Gray Scale
imLab = rgb2lab(im)
```

```python
imLab.shape
```

* `im[:,:,0]`: Intensity (L)
* `im[:,:,1]`: Green-red (A)
* `im[:,:,2]`: Blue-yellow (B)

**Gray Scale**

```python
# Lab --> Gray Scale
imGray = imLab.copy()
```

```python
imGray[:,:,1] = 0   
imGray[:,:,2] = 0   
```

```python
imGray = lab2rgb(imGray)
```

```python
plt.title('Gray Scale Image')
plt.imshow(imGray)
plt.axis('off')
plt.show()
```

**Image Brightness**

```python
# Brighter
imBrighter = imLab.copy()

# Darker
imDarker = imLab.copy()
```

```python
imBrighter[:,:,0] = imBrighter[:,:,0] + 50
```

```python
imBrighter = lab2rgb(imBrighter)
```

```python
plt.title('Brighter Image')
plt.imshow(imBrighter)
plt.axis('off')
plt.show()
```

```python
imDarker[:,:,0] = imDarker[:,:,0] - 50
```

```python
imDarker = lab2rgb(imDarker)
```

```python
plt.title('Darker Image')
plt.imshow(imDarker)
plt.axis('off')
plt.show()
```

**Invert the Image Brightness**

```python
imInvert = imLab.copy()
```

```python
imInvert[:,:,0] = np.max(imInvert[:,:,0]) - imInvert[:,:,0]
```

```python
imInvert = lab2rgb(imInvert)
```

```python
plt.title('Inverted Image')
plt.imshow(imInvert)
plt.axis('off')
plt.show()
```
